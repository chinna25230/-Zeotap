#  5 business insights from the EDA.

Insight 1: Identify the region with the highest sales and its percentage contribution



```python
region_sales = data.groupby('Region')['TotalValue'].sum().reset_index()
region_sales['Percentage'] = (region_sales['TotalValue'] / region_sales['TotalValue'].sum()) * 100
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=region_sales, x='Region', y='TotalValue', color='skyblue')
plt.title("Total Sales by Region")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.xticks(rotation=45)
plt.show()

# Pie chart for percentage contribution
plt.figure(figsize=(8, 8))
plt.pie(region_sales['Percentage'], labels=region_sales['Region'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(region_sales)))
plt.title("Percentage Contribution by Region")
plt.show()
```

Insight 2: Determine the product category with the highest revenue and average order size

```python
category_revenue = data.groupby('Category').agg(TotalRevenue=('TotalValue', 'sum'),AvgOrderSize=('Quantity', 'mean')).reset_index()

# Bar plot for category revenue
plt.figure(figsize=(8, 5))
sns.barplot(data=category_revenue, x='Category', y='TotalRevenue', hue='Category', palette='coolwarm', legend=False)
plt.title("Total Revenue by Product Category")
plt.ylabel("Total Revenue")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()

```

```python
# Bar plot for average order size
plt.figure(figsize=(8, 5))
sns.barplot(data=category_revenue, x='Category', y='AvgOrderSize', hue='Category', palette='coolwarm', legend=False)
plt.title("Average Order Size by Product Category")
plt.ylabel("Average Order Size")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.show()
```

Insight 3: Customer signup trends over time

```python
# Convert SignupDate to datetime
data['SignupDate'] = pd.to_datetime(data['SignupDate'])
data['SignupYearMonth'] = data['SignupDate'].dt.to_period('M')

# Count signups by year-month
signup_trends = data.groupby('SignupYearMonth')['CustomerID'].nunique().reset_index()
signup_trends.rename(columns={'CustomerID': 'UniqueSignups'}, inplace=True)

# Convert 'SignupYearMonth' back to datetime for proper plotting
signup_trends['SignupYearMonth'] = signup_trends['SignupYearMonth'].dt.start_time

# Line plot for signup trends
plt.figure(figsize=(10, 6))
sns.lineplot(data=signup_trends, x='SignupYearMonth', y='UniqueSignups', marker='o', color='blue')
plt.title("Customer Signup Trends Over Time")
plt.ylabel("Unique Signups")
plt.xlabel("Year-Month")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Insight 4: Top 10 customers contributing the most to revenue

```python
customer_revenue = data.groupby('CustomerName')['TotalValue'].sum().reset_index()

# Get the top 10 customers
top_customers = customer_revenue.nlargest(10, 'TotalValue')

# Bar plot for top 10 customers
plt.figure(figsize=(10, 6))
sns.barplot(data=top_customers, x='TotalValue', y='CustomerName', hue='CustomerName', palette='magma')
plt.title("Top 10 Customers by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Customer Name")
plt.show()
```

Insight 5: Product with the highest repeat purchases

```python
# Group by ProductID and count the number of unique transactions
product_repeats = data.groupby('ProductID').agg(TotalPurchases=('TransactionID', 'count'),UniqueCustomers=('CustomerID', 'nunique')).reset_index()

# Calculate repeat purchases
product_repeats['RepeatPurchases'] = product_repeats['TotalPurchases'] - product_repeats['UniqueCustomers']

# Find the top 10 products with the highest repeat purchases
top_repeat_products = product_repeats.nlargest(10, 'RepeatPurchases')

# Merge with product names for better readability
top_repeat_products = pd.merge(top_repeat_products, data[['ProductID', 'ProductName']].drop_duplicates(), on='ProductID')

# Bar plot for repeat purchases
plt.figure(figsize=(10, 6))
sns.barplot(data=top_repeat_products, x='RepeatPurchases', y='ProductName',hue='ProductName',palette='plasma')
plt.title("Top 10 Products with Highest Repeat Purchases")
plt.xlabel("Repeat Purchases")
plt.ylabel("Product Name")
plt.show()
```
