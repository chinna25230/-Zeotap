**Task 2:**

**Label encoding and min max scalar:**

**Steps:**
1. Loads the dataset and preprocesses customer data.
2. Aggregates customer transaction details into meaningful features.
3. Encodes categorical variables and normalizes numerical features.
4. Calculates cosine similarity to find the most similar customers.
5. Outputs the top 3 recommendations for the first 20 customers into a CSV file.

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Preprocessing: Aggregate customer transactional data
customer_features = data.groupby('CustomerID').agg(
    total_spend=('TotalValue','sum'),
    avg_spend=('TotalValue','mean'),
    total_quantity=('Quantity','sum'),
    unique_products=('ProductID','nunique'),
    regions=('Region','first'),
    categories=('Category',lambda x: x.mode()[0]),
).reset_index()

# Encoding categorical features
label_encoder=LabelEncoder()
customer_features['regions']=label_encoder.fit_transform(customer_features['regions'])
customer_features['categories']=label_encoder.fit_transform(customer_features['categories'])

# Normalizing numerical features for similarity calculation
scaler=MinMaxScaler()
scaled_features=scaler.fit_transform(customer_features.iloc[:,1:])

# Display the processed customer features for verification
customer_features.head(),pd.DataFrame(scaled_features, columns=customer_features.columns[1:]).head()
```

**Cosine Similarity:**

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Calculate cosine similarity between customers
similarity_matrix = cosine_similarity(scaled_features)

# Function to get top 3 similar customers for each customer
def get_top_similarities(similarity_matrix, customer_ids, top_n=3):
    lookalike_map = {}
    for idx, customer_id in enumerate(customer_ids):
        # Get similarity scores and sort by highest score excluding the customer itself
        similarity_scores = list(enumerate(similarity_matrix[idx]))
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

        # Get top N similar customers (excluding self)
        top_similar = [
            (customer_ids[i], score) for i, score in similarity_scores[1:top_n + 1]
        ]
        lookalike_map[customer_id] = top_similar
    return lookalike_map

# Get top 3 lookalikes for the first 20 customers
first_20_customers = customer_features['CustomerID'][:20].tolist()
lookalike_map = get_top_similarities(similarity_matrix, customer_features['CustomerID'].tolist())

# Extract the map for the first 20 customers
lookalike_map_20 = {cust_id: lookalike_map[cust_id] for cust_id in first_20_customers}

# Save results to "Lookalike.csv"
lookalike_df = pd.DataFrame({
    "CustomerID": lookalike_map_20.keys(),
    "Lookalikes": [str(lookalike_map_20[cust_id]) for cust_id in lookalike_map_20.keys()]
})
lookalike_csv_path = "D:\Lookalike.csv"
lookalike_df.to_csv(lookalike_csv_path, index=False)

lookalike_df.head()
```

This evaluation ensures that:
1. Ensures logic is accurate by normalizing and encoding data properly.
2. Uses cosine similarity to capture patterns effectively for high-quality recommendations.
3. Outputs the results as a CSV file for evaluation
